```python
def create_dict(map_data_file):
    cities_df = pd.read_csv(map_data_file)

    lat = list(cities_df["lat"])
    lng = list(cities_df["lng"])
    city = list(cities_df["city"])

    #print(len(set(city)))

    dict_cities = {}
    for element in range(len(city)):
        city_dict = {city[element]: (lat[element], lng[element])}
        dict_cities.update(city_dict)
        
    return dict_cities
```

```python
def slice_dict(dictionary, n):
    sliced_dict = {list(dictionary.items())[n//2]}
    return sliced_dict
```

```python
def slice_dict_left(dictionary, n):
    sliced_list = list(dictionary.items())[:n//2]

    sliced_dict = {}
    for element in sliced_list:
        sliced_dict_element = {element[0]: element[1]}
        sliced_dict.update(sliced_dict_element)

    return sliced_dict
```

```python
def slice_dict_right(dictionary, n):
    sliced_list = list(dictionary.items())[n//2 + 1:]

    sliced_dict = {}
    for element in sliced_list:
        sliced_dict_element = {element[0]: element[1]}
        sliced_dict.update(sliced_dict_element)

    return sliced_dict
```

```python
def build_kdtree(cities: dict, depth=0, k = 2):
    
    n = len(cities)

    if n <= 0:
        return None

    axis = depth % k
    
    sorted_cities = sorted(cities.items(), key=lambda cities: cities[1][axis])

    sorted_cities_dict = {}
    for element in sorted_cities:
        city_dict = {element[0]: element[1]}
        sorted_cities_dict.update(city_dict)

    return{
        'point': slice_dict(sorted_cities_dict, n),
        'left': build_kdtree(slice_dict_left(sorted_cities_dict, n), depth + 1),
        'right': build_kdtree(slice_dict_right(sorted_cities_dict, n), depth + 1)
    }
```

```python


def create_data_stucture(map_data_file):
    """Create a data structure from a data_file that allows rapid quering positions on the map

    Args:
        map_data_file (Path) csv file with the following columns:
            city,lat,lng,country,iso3,local_name,population,continent

    Returns
        data_structure that can be used with find_closest_cities

    HINT:
        BSP buildup using recursive functions
    """
    
    dict = create_dict(map_data_file)

    kd_tree = build_kdtree(dict)

    return kd_tree
```
